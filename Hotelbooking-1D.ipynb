```python
def impute_mean(series):
    return series.fillna(series.mean())
```

```python
hb.children = hb['children'].transform(impute_mean)
hb.agent = hb['agent'].transform(impute_mean)
hb.company = hb['company'].transform(impute_mean)
```

```python
hb_sort = hb.sort_values(by = 'lead_time', ascending = False)
hb_sort.head()
```

```python
def max_min_val(col):
    first = hb[col].idxmax()
    first_obs = pd.DataFrame(hb.loc[first])
    
    last = hb[col].idxmin()
    last_obs = pd.DataFrame(hb.loc[last])
    
    min_max_obs = pd.concat([first_obs, last_obs], axis=1)
    
    return min_max_obs
```

```python
hbo = hb[['agent', 'lead_time']]
hb_cat = pd.get_dummies(hbo)
cor_mat = hb_cat.corr()
mask = np.array(cor_mat)
mask[np.tril_indices_from(mask)]=False
fig = plt.gcf()
fig.set_size_inches(20,5)
sns.heatmap(data = cor_mat, mask = mask, square = True, annot = True, cbar = True);
```
